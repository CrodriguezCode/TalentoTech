# Visualizacion de Datos

```python
import matplotlib.pyplot as plt
import seaborn as sns
import plotly. express as px
import pandas as pd

# Crear un DataFrame de ejemplo
data = pd.DataFrame({
    'x': range(10),
    'y': [i**2 for i in range(10)],
    'category': ['A', 'B', 'A', 'B', 'A', 'B', 'A', 'B', 'A', 'B']
})

# Visualizacion con Matplotlib
plt.plot(data['x'], data['y'])
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Grafico de lineas con Matplotlib')
plt.show()

# Visualizacion con Seaborn
sns.scatterplot(data=data, x='x', y='y', hue='category')
plt.title('Diagrama de dispersion con Seaborn')
plt.show()

# Visualizacion con Plotly
fig = px.bar(data, x='x', y='y', color='category', barmode='group')
fig.update_layout(title='Grafico de barras con Plotly')
fig.show()
```
